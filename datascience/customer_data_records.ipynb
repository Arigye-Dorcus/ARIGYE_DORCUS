```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

```

```python
df=pd.read_csv('customer_data_records.csv')

print(df)
```

```python
df.columns
```

```python
#select features for clustering
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
#apply k-Means clustering
#use only two clusters
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(x)
```

```python
#add cluster label to dataset
df['Cluster'] = kmeans.labels_
print(df)
```

```python
#visualize the cluster
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster')
plt.title('Customer Clusters based on Income & Spending')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
z= df[['CustomerID','Annual Income (k$)']]
kmeans.fit(z)
df['cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,y='Annual Income (k$)',x='CustomerID',hue='cluster')
plt.title('customer income')
plt.ylabel('Annual Income (k$)')
plt.xlabel('CustomerID')
plt.show()
```
